```python
def calculate_percentage(part, total):
    """Calculate what percentage 'part' is of 'total'."""
    percentage = (part / total) * 100
    return round(percentage, 1)

def format_large_number(number):
    """Format large numbers with commas for readability."""
    return f"{number:,}"

Textile_waste_collected_separately_EU = 2_070_000  #  tonnes
Textile_waste_collected_separately_EU_Households = 910_000 #  tonnes
Textile_waste_collected_separately_Spain = 102_756  # tonnes
Textile_waste_collected_separately_Spain_Households = 47_631 # tonnes
Avarage_population_2022_EU = 446_847_139  # people
Avarage_population_2022_Spain = 47_786_102  # people


Spain_percentage = calculate_percentage(Textile_waste_collected_separately_Spain, Textile_waste_collected_separately_EU)
Spain_percentage_Households = calculate_percentage(Textile_waste_collected_separately_Spain_Households, Textile_waste_collected_separately_EU_Households)
Spain_percentage_population = calculate_percentage(Avarage_population_2022_Spain, Avarage_population_2022_EU)

print("EU TEXTILE WASTE STATISTICS 2022")
print("-" * 30)
print(f"Textile waste collected separately in EU: {format_large_number(Textile_waste_collected_separately_EU)}")
print(f"Textile waste collected separately in Spain: {format_large_number(Textile_waste_collected_separately_Spain)}")
print(f"Spain percentage: {Spain_percentage}%")
print(f"\nTextile waste collected separately in EU (households): {format_large_number(Textile_waste_collected_separately_EU_Households)}")
print(f"Textile waste collected separately in Spain (households): {format_large_number(Textile_waste_collected_separately_Spain_Households)}")
print(f"Spain percentage (households): {Spain_percentage_Households}%")
print(f"\nAvarage population 2022 EU: {format_large_number(Avarage_population_2022_EU)}")
print(f"Avarage population 2022 Spain: {format_large_number(Avarage_population_2022_Spain)}")
print(f"Spain percentage (population): {Spain_percentage_population}%")
```
